```python
# ========================================
# PENJELASAN ALUR ALGORITMA PROGRAM
# ========================================
# Bagian ini hanya mencetak alur kerja program sebagai panduan bagi pengguna.
print("=" * 70) # Mencetak garis pemisah untuk estetika.
print("ALUR ALGORITMA SEGMENTASI PROFIL KUALITAS AIR DENGAN K-MEANS")
print("=" * 70) # Mencetak garis pemisah untuk estetika.
print("1. IMPORT LIBRARIES & LOAD DATA")
print("   - Mengimpor library dan memuat dataset 'water_potability.csv'.")
print() # Mencetak baris kosong untuk spasi.
print("2. EXPLORATORY DATA ANALYSIS (EDA)")
print("   - Menganalisis data untuk memahami distribusi dan statistik dasarnya.")
print("   - Memvisualisasikan distribusi awal setiap fitur.")
print() # Mencetak baris kosong untuk spasi.
print("3. FEATURE SELECTION (ANALISIS KORELASI)")
print("   - Membuat heatmap untuk memeriksa korelasi antar fitur.")
print() # Mencetak baris kosong untuk spasi.
print("4. DATA PREPROCESSING")
print("   - Menangani missing values dan outliers untuk mempersiapkan data pemodelan.")
print() # Mencetak baris kosong untuk spasi.
print("5. PERSIAPAN DATA & SCALING")
print("   - Melakukan scaling pada fitur menggunakan StandardScaler.")
print() # Mencetak baris kosong untuk spasi.
print("6. MENENTUKAN JUMLAH CLUSTER OPTIMAL (K)")
print("   - Menganalisis data untuk menemukan jumlah cluster yang disarankan secara statistik.")
print("   - (Namun, akan menggunakan 3 cluster sesuai ketentuan tugas).")
print() # Mencetak baris kosong untuk spasi.
print("7. TRAINING MODEL K-MEANS DENGAN 3 CLUSTER")
print("   - Melatih model K-Means untuk mengelompokkan data ke dalam 3 profil.")
print() # Mencetak baris kosong untuk spasi.
print("8. INTERPRETASI & PEMBUATAN PROFIL 3 CLUSTER")
print("   - Menganalisis pusat massa (centroid) dari setiap cluster.")
print("   - Memvisualisasikan 3 cluster menggunakan PCA.")
print("   - Memberikan nama dan deskripsi naratif untuk setiap profil yang ditemukan.")
print("=" * 70) # Mencetak garis pemisah untuk estetika.
print() # Mencetak baris kosong untuk spasi.


# ========================================
# 1. IMPORT LIBRARIES DAN LOAD DATA
# ========================================
# Mengimpor library pandas untuk manipulasi dan analisis data, terutama untuk DataFrame.
import pandas as pd
# Mengimpor library numpy untuk operasi numerik dan array.
import numpy as np
# Mengimpor pyplot dari matplotlib untuk membuat visualisasi data seperti plot dan grafik.
import matplotlib.pyplot as plt
# Mengimpor seaborn, library visualisasi yang dibangun di atas matplotlib, untuk plot statistik yang lebih menarik.
import seaborn as sns
# Mengimpor kelas KMeans dari scikit-learn untuk menerapkan algoritma K-Means Clustering.
from sklearn.cluster import KMeans
# Mengimpor StandardScaler dari scikit-learn untuk melakukan standardisasi fitur (scaling).
from sklearn.preprocessing import StandardScaler
# Mengimpor PCA dari scikit-learn untuk reduksi dimensi, berguna untuk visualisasi.
from sklearn.decomposition import PCA
# Mengimpor silhouette_score untuk mengevaluasi kualitas hasil clustering.
from sklearn.metrics import silhouette_score
# Mengimpor library warnings untuk mengelola pesan peringatan yang mungkin muncul.
import warnings

# Mengatur agar semua pesan peringatan (warnings) diabaikan agar output lebih bersih.
warnings.filterwarnings('ignore')
# Mengatur gaya visual default untuk semua plot Matplotlib.
plt.style.use('seaborn-v0_8-darkgrid')
# Mengatur palet warna default "viridis" untuk semua plot Seaborn.
sns.set_palette("viridis")

# Mencetak header untuk bagian ini.
print("=" * 60)
print("1. MEMUAT DATASET")
print("=" * 60)

# Blok try-except untuk menangani error jika file tidak ditemukan
try:
    # Membaca file 'water_potability.csv' dan memuatnya ke dalam DataFrame pandas bernama 'df'
    df = pd.read_csv('water_potability.csv')
    # Mencetak pesan sukses jika file berhasil dimuat
    print("Dataset 'water_potability.csv' berhasil dimuat!")
    # Mencetak dimensi (jumlah baris dan kolom) dari dataset
    print(f"Shape dataset: {df.shape}")
# Menangkap error 'FileNotFoundError' jika file tidak ada di direktori
except FileNotFoundError:
    # Mencetak pesan error yang informatif
    print("ERROR: File 'water_potability.csv' tidak ditemukan.")
    print("Silakan unduh dari Kaggle dan upload ke environment Anda.")
    # Menghentikan eksekusi program jika file tidak ditemukan
    exit()


# ========================================
# 2. EXPLORATORY DATA ANALYSIS (EDA)
# ========================================
# Mencetak header untuk bagian EDA.
print("\n" + "=" * 60)
print("2. EXPLORATORY DATA ANALYSIS (EDA)")
print("=" * 60)
# Menampilkan statistik deskriptif dasar (rata-rata, std, min, max, kuartil) untuk semua kolom numerik.
print("Info dan Statistik Deskriptif Dataset:")
# .round(2) membulatkan nilai statistik menjadi 2 angka desimal.
print(df.describe().round(2))
# Menampilkan jumlah nilai yang hilang (missing values) untuk setiap kolom.
print("\nJumlah Missing Values:")
# df.isnull() mengubah setiap sel menjadi True (jika null) atau False. .sum() menjumlahkan True per kolom.
# .sort_values(ascending=False) mengurutkan dari yang memiliki missing value terbanyak.
print(df.isnull().sum().sort_values(ascending=False))

# --- [DITAMBAHKAN] Visualisasi Distribusi Fitur Awal ---
# Memberi tahu pengguna proses apa yang sedang berjalan.
print("\nMembuat visualisasi distribusi fitur awal:")
# Memilih hanya kolom yang bertipe data numerik dari DataFrame.
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
# Memeriksa apakah kolom 'Potability' (target) ada di dalam daftar kolom numerik.
if 'Potability' in numeric_cols:
    # Menghapus kolom 'Potability' dari daftar karena ini bukan fitur yang akan di-cluster.
    numeric_cols.remove('Potability')

# Membuat plot histogram untuk setiap fitur numerik yang tersisa.
# bins=20 menentukan jumlah batang pada setiap histogram.
# figsize=(15, 10) mengatur ukuran keseluruhan gambar.
# layout=(3, 3) mengatur subplot dalam grid 3x3.
df[numeric_cols].hist(bins=20, figsize=(15, 10), layout=(3, 3))
# Menambahkan judul utama untuk semua subplot.
plt.suptitle("Distribusi Setiap Fitur Sebelum Preprocessing", fontsize=16)
# Menyesuaikan layout plot secara otomatis agar tidak tumpang tindih.
plt.tight_layout(rect=[0, 0, 1, 0.96])
# Menampilkan plot yang telah dibuat.
plt.show()
# Memberi konfirmasi bahwa visualisasi telah selesai dibuat.
print("Visualisasi distribusi fitur selesai.")


# ========================================
# 3. FEATURE SELECTION (ANALISIS KORELASI)
# ========================================
# Mencetak header untuk bagian analisis korelasi.
print("\n" + "=" * 60)
print("3. FEATURE SELECTION (ANALISIS KORELASI)")
print("=" * 60)
# Membuat DataFrame baru dengan menghapus kolom 'Potability' untuk analisis antar fitur.
# errors='ignore' akan mencegah error jika kolom 'Potability' tidak ada.
features_df = df.drop('Potability', axis=1, errors='ignore')

# --- [DITAMBAHKAN] Visualisasi Heatmap Korelasi ---
# Memberi tahu pengguna proses apa yang sedang berjalan.
print("Membuat heatmap korelasi antar fitur:")
# Membuat kanvas (figure) baru untuk plot dengan ukuran 12x8 inci.
plt.figure(figsize=(12, 8))
# Menghitung matriks korelasi Pearson antar semua kolom di features_df.
correlation_matrix = features_df.corr()
# Membuat heatmap menggunakan seaborn untuk memvisualisasikan matriks korelasi.
# annot=True menampilkan nilai korelasi di setiap sel.
# cmap='viridis' mengatur skema warna.
# fmt=".2f" memformat angka menjadi 2 desimal.
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
# Menambahkan judul pada heatmap.
plt.title('Heatmap Korelasi Antar Fitur')
# Menampilkan heatmap yang telah dibuat.
plt.show()
# Mencetak hasil interpretasi dari heatmap.
print("Analisis: Tidak ada fitur dengan korelasi yang sangat tinggi")
print("sehingga semua fitur dipertahankan untuk pemodelan.")
# Memberi konfirmasi bahwa analisis telah selesai.
print("Analisis korelasi selesai.")


# ========================================
# 4. DATA PREPROCESSING
# ========================================
# Mencetak header untuk bagian preprocessing.
print("\n" + "=" * 60)
print("4. DATA PREPROCESSING")
print("=" * 60)
# Membuat salinan (copy) dari DataFrame asli agar data original tidak berubah.
df_processed = df.copy()

# Memeriksa kembali dan menghapus kolom 'Potability' dari DataFrame yang akan diproses.
if 'Potability' in df_processed.columns:
    # axis=1 menunjukkan bahwa yang dihapus adalah kolom.
    df_processed = df_processed.drop('Potability', axis=1)

# Memberi tahu pengguna langkah pertama: mengisi missing values.
print("1. Mengisi missing values dengan median...")
# Melakukan iterasi (loop) untuk setiap nama kolom di df_processed.
for col in df_processed.columns:
    # Memeriksa apakah jumlah nilai null di kolom saat ini lebih dari 0.
    if df_processed[col].isnull().sum() > 0:
        # Menghitung nilai median (nilai tengah) dari kolom tersebut.
        median_val = df_processed[col].median()
        # Mengisi semua nilai null (NaN) di kolom tersebut dengan nilai mediannya.
        # inplace=True berarti perubahan diterapkan langsung pada DataFrame.
        df_processed[col].fillna(median_val, inplace=True)
# Memberi konfirmasi bahwa proses pengisian missing values selesai.
print("Missing values telah ditangani.")

# Memberi tahu pengguna langkah kedua: menangani outliers.
print("\n2. Menangani outliers dengan metode IQR Capping...")
# Melakukan iterasi untuk setiap kolom untuk memeriksa dan menangani outliers.
for col in df_processed.columns:
    # Menghitung Kuartil 1 (persentil ke-25).
    Q1 = df_processed[col].quantile(0.25)
    # Menghitung Kuartil 3 (persentil ke-75).
    Q3 = df_processed[col].quantile(0.75)
    # Menghitung Interquartile Range (IQR).
    IQR = Q3 - Q1
    # Menghitung batas bawah untuk deteksi outlier.
    lower_bound = Q1 - 1.5 * IQR
    # Menghitung batas atas untuk deteksi outlier.
    upper_bound = Q3 + 1.5 * IQR
    # Mengganti nilai di luar batas dengan nilai batas itu sendiri (capping).
    df_processed[col] = df_processed[col].clip(lower=lower_bound, upper=upper_bound)
# Memberi konfirmasi bahwa proses penanganan outliers selesai.
print("Outliers telah ditangani.")
# Menetapkan DataFrame yang sudah bersih ke variabel X (konvensi untuk fitur).
X = df_processed
# Menyimpan nama-nama kolom fitur ke dalam sebuah list untuk referensi nanti.
feature_names = X.columns.tolist()


# ========================================
# 5. SCALING DATA
# ========================================
# Mencetak header untuk bagian scaling.
print("\n" + "=" * 60)
print("5. SCALING DATA")
print("=" * 60)
# Membuat instance (objek) dari kelas StandardScaler.
scaler = StandardScaler()
# Melakukan standardisasi pada data fitur (X).
# .fit_transform() menghitung rata-rata & std dev, lalu menerapkannya untuk mengubah data.
X_scaled = scaler.fit_transform(X)
# Memberi konfirmasi bahwa semua fitur telah di-scaling.
print("Semua fitur telah diskalakan menggunakan StandardScaler.")


# ========================================
# 6. MENENTUKAN JUMLAH CLUSTER OPTIMAL (K)
# ========================================
# Mencetak header untuk bagian penentuan K.
print("\n" + "=" * 60)
print("6. MENENTUKAN JUMLAH CLUSTER OPTIMAL (SECARA STATISTIK)")
print("=" * 60)
# Membuat rentang nilai K yang akan diuji, dari 2 sampai 10.
k_range = range(2, 11)
# Membuat list kosong untuk menyimpan nilai inertia (untuk Elbow Method).
inertias = []
# Membuat list kosong untuk menyimpan nilai silhouette score.
silhouette_scores = []

# Melakukan iterasi untuk setiap nilai K dalam rentang yang ditentukan.
for k in k_range:
    # Membuat model KMeans dengan jumlah cluster 'k'.
    # random_state=42 untuk hasil yang konsisten, n_init=10 untuk stabilitas model.
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    # Melatih model K-Means pada data yang sudah di-scaling.
    kmeans.fit(X_scaled)
    # Menghitung inertia (total jarak kuadrat sampel ke pusat clusternya) dan menambahkannya ke list.
    inertias.append(kmeans.inertia_)
    # Menghitung silhouette score dan menambahkannya ke list.
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Membuat figure dan 2 subplot (axes) berdampingan dengan ukuran total 16x6 inci.
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
# Membuat plot untuk Elbow Method pada subplot pertama (kiri).
axes[0].plot(k_range, inertias, 'bo-')
# Memberi label sumbu-x.
axes[0].set_xlabel('Jumlah Cluster (k)')
# Memberi label sumbu-y.
axes[0].set_ylabel('Inertia')
# Memberi judul pada subplot.
axes[0].set_title('Elbow Method')
# Membuat plot untuk Silhouette Score pada subplot kedua (kanan).
axes[1].plot(k_range, silhouette_scores, 'ro-')
# Memberi label sumbu-x.
axes[1].set_xlabel('Jumlah Cluster (k)')
# Memberi label sumbu-y.
axes[1].set_ylabel('Silhouette Score')
# Memberi judul pada subplot.
axes[1].set_title('Silhouette Score')
# Menampilkan kedua plot.
plt.show()


# ========================================
# 7. TRAINING MODEL K-MEANS
# ========================================
# Mencetak header untuk bagian training.
print("\n" + "=" * 60)
print(f"7. TRAINING MODEL K-MEANS DENGAN K={jumlah_cluster_tugas}")
print("=" * 60)
# Menginisialisasi model K-Means final dengan 3 cluster.
kmeans_model = KMeans(n_clusters=jumlah_cluster_tugas, random_state=42, n_init=10)
# Melatih model dan sekaligus memprediksi label cluster untuk setiap sampel data.
clusters = kmeans_model.fit_predict(X_scaled)
# Memberi konfirmasi bahwa model telah dilatih.
print("Model K-Means berhasil dilatih dan data telah dikelompokkan.")
# Menambahkan kolom 'Cluster' baru ke DataFrame X yang berisi label cluster hasil prediksi.
X['Cluster'] = clusters


# ========================================
# 8. INTERPRETASI & PEMBUATAN PROFIL CLUSTER
# ========================================
# Mencetak header untuk bagian interpretasi.
print("\n" + "=" * 60)
print("8. INTERPRETASI PROFIL SETIAP CLUSTER")
print("=" * 60)

# Memberi judul untuk sub-bagian pertama.
print("1. Distribusi Ukuran Setiap Cluster:")
# Menghitung jumlah sampel (baris) untuk setiap ID cluster (0, 1, 2).
# .sort_index() memastikan urutan cluster benar (0, 1, 2).
cluster_counts = X['Cluster'].value_counts().sort_index()
# Membuat kanvas baru untuk plot bar.
plt.figure(figsize=(10, 6))
# Membuat plot batang untuk memvisualisasikan jumlah sampel per cluster.
sns.barplot(x=cluster_counts.index, y=cluster_counts.values, palette="viridis")
# Memberi judul pada plot.
plt.title('Jumlah Sampel per Cluster')
# Memberi label sumbu-x.
plt.xlabel('Cluster ID')
# Memberi label sumbu-y.
plt.ylabel('Jumlah Sampel Air')
# Menampilkan plot.
plt.show()
# Mencetak nilai jumlah sampel per cluster dalam bentuk teks.
print(cluster_counts)

# Memberi judul untuk sub-bagian kedua.
print("\n2. Analisis Pusat Cluster (Centroids):")
# Memberi penjelasan tentang tabel yang akan ditampilkan.
print("   Tabel di bawah ini menunjukkan nilai rata-rata setiap fitur untuk masing-masing cluster.")
# Mengubah nilai centroid yang tadinya di-scaling kembali ke skala data aslinya agar mudah diinterpretasi.
centroids_original_scale = scaler.inverse_transform(kmeans_model.cluster_centers_)
# Membuat DataFrame dari nilai-nilai centroid agar mudah dibaca.
centroid_df = pd.DataFrame(centroids_original_scale, columns=feature_names)
# Mencetak DataFrame centroid dengan nilai dibulatkan 2 desimal.
print(centroid_df.round(2))

# Memberi judul untuk sub-bagian ketiga.
print("\n3. Visualisasi Cluster dalam Ruang 2D (menggunakan PCA):")
# Membuat instance PCA untuk mereduksi data menjadi 2 komponen utama.
pca = PCA(n_components=2)
# Menerapkan PCA pada data yang sudah di-scaling untuk mendapatkan representasi 2D.
X_pca = pca.fit_transform(X_scaled)
# Membuat kanvas baru untuk scatter plot.
plt.figure(figsize=(12, 8))
# Membuat scatter plot dari data 2D hasil PCA.
# hue=clusters memberi warna pada titik berdasarkan label clusternya.
# s=50 mengatur ukuran titik, alpha=0.7 mengatur transparansi.
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette="viridis", s=50, alpha=0.7)
# Memberi judul pada plot.
plt.title('Visualisasi Cluster Kualitas Air dengan PCA')
# Memberi label sumbu-x, termasuk informasi varians yang dijelaskan oleh Komponen Utama 1.
plt.xlabel(f'Principal Component 1 ({pca.explained_variance_ratio_[0]*100:.2f}%)')
# Memberi label sumbu-y, termasuk informasi varians yang dijelaskan oleh Komponen Utama 2.
plt.ylabel(f'Principal Component 2 ({pca.explained_variance_ratio_[1]*100:.2f}%)')
# Menambahkan legenda untuk ID cluster.
plt.legend(title='Cluster ID')
# Menambahkan grid pada plot untuk kemudahan membaca.
plt.grid(True)
# Menampilkan plot.
plt.show()

# Memberi judul untuk sub-bagian keempat.
print("\n4. Deskripsi Profil untuk Setiap Cluster:")
# Melakukan iterasi untuk setiap ID cluster (0, 1, dan 2).
for i in range(jumlah_cluster_tugas):
    # Mencetak garis pemisah untuk setiap profil cluster.
    print("-" * 40)
    # Mencetak judul profil, termasuk ID cluster dan jumlah anggotanya.
    print(f"AIR = PROFIL CLUSTER {i} (Total: {cluster_counts[i]} sampel)")
    # Mencetak garis pemisah.
    print("-" * 40)
    # Menghitung rata-rata keseluruhan untuk setiap fitur dari semua cluster.
    overall_mean = centroid_df.mean()
    # Mengambil nilai centroid untuk cluster saat ini (cluster i).
    cluster_centroid = centroid_df.loc[i]
    # Menghitung seberapa jauh nilai centroid cluster saat ini dari rata-rata keseluruhan.
    deviation = cluster_centroid - overall_mean
    # Menemukan 3 fitur yang nilainya paling tinggi (paling menonjol) di cluster ini.
    top_features = deviation.nlargest(3)
    # Menemukan 2 fitur yang nilainya paling rendah di cluster ini.
    bottom_features = deviation.nsmallest(2)
    # Mencetak header untuk karakteristik utama.
    print("Karakteristik Utama:")
    # Mencetak fitur-fitur yang menonjol tinggi.
    print("   - Menonjol Tinggi pada:")
    # Melakukan iterasi pada fitur-fitur yang paling menonjol tinggi.
    for feature, val in top_features.items():
        # Mencetak nama fitur dan nilainya.
        print(f"     - {feature}: {cluster_centroid[feature]:.2f} (di atas rata-rata)")
    # Mencetak fitur-fitur yang menonjol rendah.
    print("   - Menonjol Rendah pada:")
    # Melakukan iterasi pada fitur-fitur yang paling menonjol rendah.
    for feature, val in bottom_features.items():
        # Mencetak nama fitur dan nilainya.
        print(f"     - {feature}: {cluster_centroid[feature]:.2f} (di bawah rata-rata)")

# --- KESIMPULAN AKHIR ---
# Mencetak header untuk kesimpulan akhir.
print("\n" + "=" * 60)
print("KESIMPULAN AKHIR")
print("=" * 60)
# Mencetak kesimpulan dari analisis clustering.
print(f"Analisis clustering berhasil mengidentifikasi {jumlah_cluster_tugas} profil kualitas air yang berbeda dalam dataset.")
print("Setiap profil memiliki karakteristik kimia yang unik, yang dapat berguna untuk:")
# Mencetak poin-poin kegunaan hasil analisis.
print("   - Memahami variasi sumber air baku.")
print("   - Merancang strategi pengolahan air yang berbeda untuk setiap tipe air.")
print("   - Sebagai langkah awal untuk analisis lebih lanjut seperti deteksi anomali.")
# Menekankan sifat unsupervised dari analisis ini.
print("\nAnalisis murni unsupervised ini memberikan wawasan tentang struktur internal data tanpa memerlukan label apapun.")
```
