```python
# ========================================
# PENJELASAN ALUR ALGORITMA PROGRAM
# ========================================
# Bagian ini hanya mencetak alur kerja program ke layar sebagai panduan bagi pengguna.
print("=" * 70) # Mencetak garis pemisah untuk kerapian output.
print("ALUR ALGORITMA PREDIKSI POTABILITAS AIR") # Mencetak judul utama program.
print("=" * 70) # Mencetak garis pemisah.
print("1. IMPORT LIBRARIES & LOAD DATA") # Menjelaskan langkah pertama.
print("   - Import semua library yang diperlukan untuk analisis") # Detail langkah.
print("   - Load dataset water_potability.csv dengan error handling") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("2. EXPLORATORY DATA ANALYSIS (EDA)") # Menjelaskan langkah kedua.
print("   - Analisis informasi dasar dataset (shape, info, statistik)") # Detail langkah.
print("   - Deteksi missing values pada setiap kolom") # Detail langkah.
print("   - Visualisasi distribusi target variable (Potability)") # Detail langkah.
print("   - Analisis korelasi antar fitur menggunakan heatmap") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("3. DATA PREPROCESSING") # Menjelaskan langkah ketiga.
print("   - Pisahkan fitur (X) dan target variable (y)") # Detail langkah.
print("   - Handle missing values menggunakan SimpleImputer (median)") # Detail langkah.
print("   - Feature scaling menggunakan StandardScaler") # Detail langkah.
print("   - Normalisasi penting untuk Logistic Regression") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("4. SPLIT DATA TRAINING & TESTING") # Menjelaskan langkah keempat.
print("   - Bagi data menjadi 80% training dan 20% testing") # Detail langkah.
print("   - Gunakan stratified split untuk menjaga distribusi target") # Detail langkah.
print("   - Stratify=y memastikan proporsi kelas seimbang") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("5. IMPLEMENTASI MODEL LOGISTIC REGRESSION") # Menjelaskan langkah kelima.
print("   - Inisialisasi model Logistic Regression") # Detail langkah.
print("   - Evaluasi menggunakan 5-fold Cross Validation") # Detail langkah.
print("   - Training model pada data training") # Detail langkah.
print("   - Prediksi pada data testing") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("6. EVALUASI PERFORMA MODEL") # Menjelaskan langkah keenam.
print("   - Confusion Matrix dengan visualisasi heatmap") # Detail langkah.
print("   - Classification Report (precision, recall, f1-score)") # Detail langkah.
print("   - Metrik evaluasi: Accuracy, Precision, Recall, F1, AUC-ROC") # Detail langkah.
print("   - Interpretasi hasil evaluasi") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("7. INTERPRETASI HASIL & ANALISIS KOEFISIEN") # Menjelaskan langkah ketujuh.
print("   - Analisis koefisien model untuk interpretasi fitur") # Detail langkah.
print("   - Identifikasi fitur yang paling berpengaruh") # Detail langkah.
print("   - Kesimpulan dan rekomendasi pengembangan") # Detail langkah.
print() # Mencetak baris kosong sebagai spasi.
print("METODE SUPERVISED LEARNING: Logistic Regression") # Menyebutkan metode yang digunakan.
print("TASK: Binary Classification (Layak/Tidak Layak Minum)") # Menyebutkan jenis tugas machine learning.
print("EVALUASI: Cross Validation, Confusion Matrix, Multiple Metrics") # Menyebutkan teknik evaluasi.
print("=" * 70) # Mencetak garis pemisah.
print() # Mencetak baris kosong sebagai spasi.


# ========================================
# 1. IMPORT LIBRARIES DAN LOAD DATA
# ========================================
# Mengimpor library pandas untuk manipulasi data (membaca CSV, membuat DataFrame).
import pandas as pd
# Mengimpor library numpy untuk operasi numerik, terutama array.
import numpy as np
# Mengimpor matplotlib untuk membuat visualisasi/grafik dasar.
import matplotlib.pyplot as plt
# Mengimpor seaborn untuk membuat visualisasi statistik yang lebih menarik dan informatif.
import seaborn as sns
# Mengimpor fungsi-fungsi spesifik dari scikit-learn untuk pemodelan machine learning.
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold # Untuk membagi data, validasi silang, dan stratifikasi.
from sklearn.preprocessing import StandardScaler # Untuk melakukan penskalaan fitur (standardisasi).
from sklearn.impute import SimpleImputer # Untuk menangani nilai yang hilang (missing values).
from sklearn.linear_model import LogisticRegression # Untuk model klasifikasi Regresi Logistik.
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score # Untuk evaluasi performa model.
# Mengimpor library warnings dan mengaturnya untuk mengabaikan pesan peringatan agar output lebih bersih.
import warnings
warnings.filterwarnings('ignore') # Mengabaikan pesan peringatan yang mungkin muncul saat menjalankan kode.

print("=" * 60) # Mencetak garis pemisah.
print("ANALISIS PREDIKSI POTABILITAS AIR DENGAN REGRESI LOGISTIK") # Mencetak judul analisis.
print("=" * 60) # Mencetak garis pemisah.

# Blok try-except digunakan untuk menangani error secara elegan jika file tidak ditemukan.
try:
    # Membaca file 'water_potability.csv' dan memuatnya ke dalam DataFrame pandas bernama 'df'.
    df = pd.read_csv('water_potability.csv')
    # Mencetak pesan sukses jika file berhasil ditemukan dan dimuat.
    print("Dataset 'water_potability.csv' berhasil dimuat!")
    # Mencetak dimensi (jumlah baris dan kolom) dari dataset menggunakan atribut .shape.
    print(f"Shape dataset: {df.shape}")
# Menangkap error 'FileNotFoundError' jika file tidak ada di direktori yang sama dengan script.
except FileNotFoundError:
    # Mencetak pesan error yang informatif kepada pengguna.
    print("ERROR: File 'water_potability.csv' tidak ditemukan.")
    print("Silakan unduh dari Kaggle dan upload ke environment Anda.")
    # Menghentikan eksekusi program jika file tidak ditemukan karena analisis tidak dapat dilanjutkan.
    exit()

# ========================================
# 2. EXPLORATORY DATA ANALYSIS (EDA)
# ========================================
print("\n" + "=" * 50) # Mencetak judul bagian EDA.
print("2. EXPLORATORY DATA ANALYSIS (EDA)")
print("=" * 50)

# Menampilkan ringkasan informasi DataFrame menggunakan metode .info().
# Ini mencakup jumlah entri, tipe data setiap kolom, dan jumlah nilai non-null.
print("Info Dataset:")
df.info()
# Menampilkan statistik deskriptif untuk semua kolom numerik menggunakan metode .describe().
# Ini termasuk hitungan, rata-rata, standar deviasi, nilai minimum, kuartil, dan nilai maksimum.
print("\nStatistik Deskriptif:")
print(df.describe())

# Memeriksa setiap sel di DataFrame apakah nilainya null/kosong (True/False), lalu menjumlahkan nilai True untuk setiap kolom.
print("\nJumlah Missing Values per Kolom:")
print(df.isnull().sum())

# Membuat kanvas/figure baru untuk plot dengan ukuran (lebar, tinggi) 6x5 inci.
plt.figure(figsize=(6, 5))
# Membuat plot batang (bar plot) untuk menghitung jumlah setiap nilai unik di kolom 'Potability'.
sns.countplot(data=df, x='Potability')
# Memberikan judul pada plot agar mudah dipahami.
plt.title('Distribusi Target: Potability (0 = Tidak Layak, 1 = Layak Minum)')
# Menampilkan plot yang sudah dibuat ke layar.
plt.show()

# Membuat figure baru untuk heatmap dengan ukuran 12x10 inci agar lebih mudah dibaca.
plt.figure(figsize=(12, 10))
# Membuat heatmap dari matriks korelasi antar fitur.
# df.corr() menghitung korelasi Pearson antar kolom.
# annot=True menampilkan nilai korelasi di setiap sel heatmap.
# cmap='coolwarm' mengatur skema warna (biru untuk negatif, merah untuk positif).
# fmt='.2f' memformat angka menjadi dua desimal.
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
# Memberikan judul pada heatmap.
plt.title('Heatmap Korelasi Fitur')
# Menampilkan heatmap yang sudah dibuat.
plt.show()

# ========================================
# 3. DATA PREPROCESSING
# ========================================
print("\n" + "=" * 50) # Mencetak judul bagian Preprocessing.
print("3. DATA PREPROCESSING")
print("=" * 50)

# Memisahkan fitur (variabel independen, X) dari target (variabel dependen, y).
# X berisi semua kolom dari DataFrame KECUALI kolom 'Potability'. axis=1 berarti menghapus kolom.
X = df.drop('Potability', axis=1)
# y hanya berisi kolom target, yaitu 'Potability'.
y = df['Potability']

# --- Handling Missing Values ---
# Membuat objek (instance) dari kelas SimpleImputer.
# Strategi 'median' dipilih untuk mengisi nilai yang hilang dengan nilai median dari masing-masing kolom.
imputer = SimpleImputer(strategy='median')
# Menghitung median dari setiap kolom di X (fit) dan sekaligus mengisi semua nilai NaN dengan median tersebut (transform).
X_imputed = imputer.fit_transform(X)
# Mengubah hasil imputasi (yang berupa array numpy) kembali menjadi DataFrame pandas agar struktur dan nama kolom tetap terjaga.
X_imputed = pd.DataFrame(X_imputed, columns=X.columns)
print("Missing values telah diisi menggunakan median.")

# --- Feature Scaling ---
# Membuat objek dari kelas StandardScaler untuk menstandarisasi fitur.
# Standardisasi mengubah data sehingga memiliki rata-rata 0 dan standar deviasi 1.
scaler = StandardScaler()
# Menghitung rata-rata dan standar deviasi dari X_imputed (fit) lalu menerapkan transformasi penskalaan (transform).
X_scaled = scaler.fit_transform(X_imputed)
# Mengubah hasil penskalaan (array numpy) kembali menjadi DataFrame pandas.
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print("Fitur telah diskalakan menggunakan StandardScaler.")

# ========================================
# 4. SPLIT DATA TRAINING DAN TESTING
# ========================================
print("\n" + "=" * 50) # Mencetak judul bagian pembagian data.
print("4. PEMBAGIAN DATA TRAINING & TESTING")
print("=" * 50)

# Membagi data menjadi set data training dan data testing menggunakan fungsi train_test_split.
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y,                 # Data fitur (X_scaled) dan target (y) yang akan dibagi.
    test_size=0.2,               # Proporsi data yang akan dialokasikan untuk testing (20%). Sisanya (80%) untuk training.
    random_state=42,             # Menjamin bahwa pembagian data akan selalu sama setiap kali kode dijalankan (reproducibility).
    stratify=y                   # Menjaga agar proporsi kelas target (jumlah 0 dan 1) tetap sama di data train dan test, penting untuk data tidak seimbang.
)
# Mencetak jumlah sampel (baris) dalam data training.
print(f"Ukuran Data Training: {X_train.shape[0]} sampel")
# Mencetak jumlah sampel (baris) dalam data testing.
print(f"Ukuran Data Testing: {X_test.shape[0]} sampel")
# Memberikan konfirmasi bahwa stratifikasi berhasil dilakukan.
print(f"Distribusi target di data training terjaga berkat 'stratify=y'.")

# ========================================
# 5. IMPLEMENTASI & EVALUASI MODEL REGRESI LOGISTIK
# ========================================
print("\n" + "=" * 50) # Mencetak judul bagian pemodelan.
print("5. PEMODELAN DENGAN REGRESI LOGISTIK")
print("=" * 50)

# Menginisialisasi model Regresi Logistik.
# random_state=42 untuk hasil yang konsisten, max_iter=1000 untuk memberikan lebih banyak iterasi bagi solver untuk konvergen.
model = LogisticRegression(random_state=42, max_iter=1000)

# --- Cross Validation ---
# Mendefinisikan strategi validasi silang (5-fold) yang juga menjaga stratifikasi.
# n_splits=5 berarti data akan dibagi menjadi 5 bagian (fold). shuffle=True mengacak data sebelum dibagi.
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
# Melakukan 5-fold cross-validation pada data training dan menghitung skor akurasi untuk setiap fold.
cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy')

# Mencetak rata-rata dan standar deviasi (dikalikan 2) dari skor cross-validation.
# .mean() menghitung rata-rata, .std() menghitung standar deviasi.
# :.4f memformat angka menjadi 4 desimal.
print(f"Akurasi Cross-Validation (5-fold): {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
print("Ini menunjukkan seberapa baik model akan berperforma pada data baru secara umum.")

# --- Training Model ---
# Melatih (fitting) model Regresi Logistik menggunakan seluruh data training (X_train, y_train).
model.fit(X_train, y_train)
print("\n Model Regresi Logistik berhasil di-training.")

# --- Prediksi pada Data Testing ---
# Menggunakan model yang sudah dilatih untuk memprediksi label kelas pada data test (X_test).
y_pred = model.predict(X_test)
# Memprediksi probabilitas keanggotaan setiap kelas.
# [:, 1] mengambil probabilitas untuk kelas positif (kelas 1, yaitu 'Potable'). Ini diperlukan untuk metrik AUC-ROC.
y_pred_proba = model.predict_proba(X_test)[:, 1]

# ========================================
# 6. EVALUASI PERFORMA MODEL
# ========================================
print("\n" + "=" * 50) # Mencetak judul bagian evaluasi.
print("6. EVALUASI PERFORMA MODEL PADA DATA TESTING")
print("=" * 50)

# --- Confusion Matrix ---
# Menghitung confusion matrix dengan membandingkan nilai aktual dari data test (y_test) dan nilai hasil prediksi model (y_pred).
cm = confusion_matrix(y_test, y_pred)
# Membuat figure baru untuk heatmap dengan ukuran 8x6 inci.
plt.figure(figsize=(8, 6))
# Memvisualisasikan confusion matrix menggunakan heatmap dari seaborn untuk interpretasi yang lebih mudah.
# annot=True menampilkan angka di setiap sel. fmt='d' memformat angka sebagai integer. cmap='Blues' menggunakan skema warna biru.
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Prediksi Tidak Layak', 'Prediksi Layak'], # Label untuk sumbu x (prediksi).
            yticklabels=['Aktual Tidak Layak', 'Aktual Layak'])     # Label untuk sumbu y (aktual).
plt.title('Confusion Matrix - Regresi Logistik') # Judul plot.
plt.ylabel('Label Aktual') # Label sumbu y.
plt.xlabel('Label Prediksi') # Label sumbu x.
plt.show() # Menampilkan plot.

# Mencetak interpretasi dari setiap sel di confusion matrix untuk memperjelas hasilnya.
print("\n--- Penjelasan Confusion Matrix ---")
print(f"True Negative (TN): {cm[0][0]} -> Air tidak layak & diprediksi tidak layak (Benar)")
print(f"False Positive (FP): {cm[0][1]} -> Air tidak layak & diprediksi layak (Salah - Tipe Error I, Berbahaya!)")
print(f"False Negative (FN): {cm[1][0]} -> Air layak & diprediksi tidak layak (Salah - Tipe Error II, Kurang Efisien)")
print(f"True Positive (TP): {cm[1][1]} -> Air layak & diprediksi layak (Benar)")

# --- Classification Report & Key Metrics ---
# Mencetak laporan klasifikasi yang berisi precision, recall, f1-score untuk setiap kelas ('Tidak Layak' dan 'Layak').
print("\n--- Laporan Klasifikasi ---")
print(classification_report(y_test, y_pred, target_names=['Tidak Layak (0)', 'Layak (1)']))

# Menghitung metrik performa utama secara terpisah untuk analisis lebih lanjut.
print("--- Metrik Kunci ---")
accuracy = accuracy_score(y_test, y_pred) # Akurasi: (TP + TN) / Total
precision = precision_score(y_test, y_pred) # Presisi: TP / (TP + FP)
recall = recall_score(y_test, y_pred) # Recall (Sensitivitas): TP / (TP + FN)
f1 = f1_score(y_test, y_pred) # F1-Score: Rata-rata harmonik dari presisi dan recall.
auc = roc_auc_score(y_test, y_pred_proba) # AUC-ROC: Area di bawah kurva ROC.

# Mencetak setiap metrik beserta penjelasannya yang sederhana dan mudah dipahami.
print(f"Akurasi: {accuracy:.4f} -> (Proporsi prediksi yang benar secara keseluruhan)")
print(f"Presisi: {precision:.4f} -> (Dari semua yang diprediksi 'Layak', berapa persen yang benar-benar layak?)")
print(f"Recall (Sensitivitas): {recall:.4f} -> (Dari semua air yang sebenarnya 'Layak', berapa persen yang berhasil terdeteksi?)")
print(f"F1-Score: {f1:.4f} -> (Rata-rata harmonik dari Presisi dan Recall)")
print(f"AUC-ROC: {auc:.4f} -> (Kemampuan model membedakan antara kelas positif dan negatif)")

# ========================================
# 7. ANALISIS DAN INTERPRETASI HASIL
# ========================================
print("\n" + "=" * 50) # Mencetak judul bagian interpretasi.
print("7. INTERPRETASI HASIL MODEL")
print("=" * 50)

# --- Interpretasi Koefisien Model ---
# Mengambil koefisien (bobot) yang dipelajari model untuk setiap fitur dari atribut 'model.coef_'.
# model.coef_ adalah array 2D, jadi [0] digunakan untuk mengakses array koefisien pertama.
# Kemudian menyusunnya dalam DataFrame pandas agar mudah dibaca dan diurutkan.
coefficients = pd.DataFrame({
    'Fitur': X.columns, # Kolom pertama berisi nama-nama fitur.
    'Koefisien': model.coef_[0] # Kolom kedua berisi nilai koefisien untuk setiap fitur.
}).sort_values(by='Koefisien', ascending=False) # Diurutkan dari yang koefisiennya paling tinggi (paling positif) ke paling rendah.

# Mencetak tabel koefisien yang sudah diurutkan.
print("Pengaruh Setiap Fitur terhadap Prediksi 'Layak Minum':")
print(coefficients)
# Mencetak penjelasan cara membaca nilai koefisien.
print("\nPenjelasan:")
print("- Koefisien POSITIF: Peningkatan nilai fitur ini berhubungan dengan peningkatan kemungkinan air LAYAK minum.")
print("- Koefisien NEGATIF: Peningkatan nilai fitur ini berhubungan dengan penurunan kemungkinan air LAYAK minum (cenderung TIDAK LAYAK).")

# --- KESIMPULAN AKHIR ---
# Mencetak ringkasan akhir dari hasil analisis dalam bahasa yang mudah dipahami.
print("\n--- KESIMPULAN AKHIR ---")
# Menampilkan akurasi akhir dalam bentuk persentase.
print(f"Model Regresi Logistik yang dibangun memberikan akurasi sebesar {accuracy:.2%} pada data testing.")
# Memberikan catatan kritis terhadap metrik lain selain akurasi.
print("Meskipun akurasinya cukup baik, nilai Presisi dan Recall perlu diperhatikan.")
# Mengidentifikasi fitur paling berpengaruh berdasarkan tabel koefisien. .iloc[0,0] mengambil nama fitur dengan koefisien tertinggi. .iloc[-1,0] yang terendah.
print(f"Model ini menunjukkan bahwa fitur seperti '{coefficients.iloc[0,0]}' memiliki pengaruh positif terbesar, sementara '{coefficients.iloc[-1,0]}' memiliki pengaruh negatif terbesar terhadap potabilitas air.")
# Memberikan rekomendasi untuk pengembangan model lebih lanjut.
print("\nModel ini dapat menjadi alat bantu awal yang baik, namun untuk aplikasi nyata, peningkatan model lebih lanjut (misalnya dengan hyperparameter tuning atau menggunakan model yang lebih kompleks) mungkin diperlukan untuk meningkatkan presisi dalam mendeteksi air yang tidak layak minum (menghindari False Positives).")

# Menandai akhir dari seluruh proses analisis.
print("\n" + "=" * 60)
print("ANALISIS SELESAI")
print("=" * 60)
```
